# Chapter 4 Exercise 6:Suppose we collect data for a group of students in a statistics class with variables X1 =hours studied, X2 =undergrad GPA, and Y = receive an A. We fit a logistic regression and produce estimated coefficient, ˆ β0 = −6, ˆβ1 = 0.05, ˆβ2 = 1.
Exercise 6A:Estimate the probability that a student who studies for 40 h and has an undergrad GPA of 3.5 gets an A in the class.

Estimate probability with this equation 
P(X) = e(β0 + β1X1 + β2X2)/(1 + e(β0 + β1X1 + β2X2))

Substituting the variables 
p(X)= e(−6+0.05∗40+1∗3.5)/(1+e(−6+0.05∗40+1∗3.5))

A student who studies for 40h and has a GPA of 3.5 will get an A by 0.3775

```r
#Exercise 10B:Use the full data set to perform a logistic regression with Direction as the response and the five lag variables plus Volume
#as predictors. Use the summary function to print the results.

#Get full model
glm_fit=glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume, data=Weekly,family = binomial)
summary(glm_fit)
#From the observation, Lag 2 is significantly important with significance of 3%
```

```r
#Exercise 10D:Now fit the logistic regression model using a training data period from 1990 to 2008, with Lag2 as the only predictor.
train <- Weekly[Weekly$Year <= 2008, ]
test <- Weekly[Weekly$Year > 2008, ]
glmdirect <- glm(Direction ~ Lag2, data = train, family = "binomial")
prediction <- factor(ifelse(predict(glmdirect, newdata = test, type = "response") < 0.5, "Down", "Up"))
confusionMatrix(prediction, test$Direction, positive = "Up")
#The accuracy is 0.625 or 62.5% and Pvalue of 0.2439 >0.05, we also have a 58.65% in the no information rate which gives us more insight to the largest part of the test dataset is our baseline classifier.
```

# Chapter 4 Exercise 11:In this problem, you will develop a model to predict whether a given car gets high or low gas mileage based on the Auto data set.

```r
#Exercise 11A:Create a binary variable, mpg01, that contains a 1 if mpg contains a value above its median, and a 0 if mpg contains a value below its median. 
mpg01 <- rep(0, length(mpg))
mpg01[mpg > median(mpg)] <- 1
Auto <- data.frame(Auto, mpg01)
```

```r
#Exercise 11B:Explore the data graphically in order to investigate the association between mpg01 and the other features.
Auto$name <- NULL
Auto$mpg <- NULL
Auto$origin <- factor(Auto$origin, labels = c("American", "European", "Japanese"))

#printing out result in transposed form
glimpse(Auto)

#Check the relationship
pairs(Auto)

#cylinders, displacement, horsepower, and weight are in an inversely associated with mpg01.
```

```r
#Exercise 11C:Split the data into a training set and a test set.
#We use a 70-30 split where 70% will be used for Training the model and 30% for testing the model.
set.seed(1)
subset <- sample(nrow(Auto), nrow(Auto) * 0.7)
trainauto = Auto[subset, ]
testauto = Auto[-subset, ]
dim(trainauto)
dim(testauto)
```

```r
#Exercise 11D:Perform LDA on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?
set.seed(1)
trainauto.lda.fit<-lda(mpg01~cylinders+displacement+horsepower+weight,data=trainauto)
testauto.lda.predict=predict(trainauto.lda.fit, testauto)
testauto.lda.class= testauto.lda.predict$class
table(testauto$mpg01, testauto.lda.class,dnn = c("Actual Mileage", "Predicted Mileage"))

round(mean(testauto.lda.class!= testauto$mpg01),3)
#The LDA Test error for the model is 0.119
```

```r
#Exercise 11E:Perform QDA on the training data in order to predict mpg01 using the variables that seemed most associated with mpg01 in (b). What is the test error of the model obtained?
set.seed(1)
trainauto.qda.fit<-qda(mpg01~cylinders+displacement+horsepower+weight,data=trainauto)
testauto.qda.predict=predict(trainauto.qda.fit,testauto)
testauto.qda.class=testauto.qda.predict$class
table(testauto$mpg01,testauto.qda.class,dnn = c("Actual Mileage", "Predicted Mileage"))    
round(mean(testauto.qda.class!= testauto$mpg01),3)    
#LDA performs better than the QDA in terms of test error. The QDA test error is  0.119
```

```r
#Exercise 11F:Perform logistic regression on the training data in order to predict mpg01 using the variables that seemed most associated with pg01 in (b). What is the test error of the model obtained?
set.seed(1)
trainauto.glm.fit<-glm(mpg01~cylinders+displacement+horsepower+weight,data=trainauto,family = binomial)
testauto.glm.predict=predict(trainauto.glm.fit,testauto,type="response")
testauto.glm.class=ifelse(testauto.glm.predict>0.5,1,0)
table(testauto$mpg01,testauto.glm.class,dnn = c("Actual Mileage", "Predicted Mileage"))
round(mean(testauto.glm.class!=testauto$mpg01),3)
#Test error for logistic regression is 0.093 for the model for the variables: horsepower, weight, cylinder, displacement
```

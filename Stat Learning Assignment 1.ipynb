```r
#Chapter 3 Question 15
#Question 15A. For each predictor, fit a simple linear regression model to predict the response.
bostonzn<-lm(crim ~ zn, data=Boston)
summary(bostonzn)

bostonindus<-lm(crim ~ indus, data=Boston)
summary(bostonindus)

bostonchas<-lm(crim ~ chas, data=Boston)
summary(bostonchas)

bostonnox<-lm(crim ~ nox, data=Boston)
summary(bostonnox)

bostonrm<-lm(crim ~ rm, data=Boston)
summary(bostonrm)

bostonage<-lm(crim ~ age, data=Boston)
summary(bostonage)

bostondis<-lm(crim ~ dis, data=Boston)
summary(bostondis)

bostonrad<-lm(crim ~ rad, data=Boston)
summary(bostonrad)

bostontax<-lm(crim ~ tax, data=Boston)
summary(bostontax)

bostonptratio<-lm(crim ~ ptratio, data=Boston)
summary(bostonptratio)

bostonblack<-lm(crim ~ black, data=Boston)
summary(bostonblack)

bostonlstat<-lm(crim ~ lstat, data=Boston)
summary(bostonlstat)

bostonmedv<-lm(crim ~ medv, data=Boston)
summary(bostonmedv)

#Only the chas variable is not important in predicting the per capita crime rate, according to the models above.
#We could also reject null hypothesis based on the p-value of its t statistic. 
#We can reject null hypothesis and conclude that there is a statistically significant relationship between predictor and response since the p-value for other parameter is too small

#Question 15B. Fit a multiple regression model to predict the response using all of the predictors.
bostonall<-lm(crim ~ ., data=Boston)
summary(bostonall)

#From the analysis above we can reject null hypotheses for variables:zn, dis, rad, black, medv,whose p-value is lower than 0.05

#Question 15C. How do your results from (a) compare to your results from (b)? Create a plot displaying the univariate regression coefficients from (a) on the x-axis, and the multiple regression coefficients from (b) on the y-axis
result<-vector("numeric",0)
result<-c(result,bostonzn$coefficients[2])
result<-c(result,bostonindus$coefficients[2])
result<-c(result,bostonchas$coefficients[2])
result<-c(result,bostonnox$coefficients[2])
result<-c(result,bostonrm$coefficients[2])
result<-c(result,bostonage$coefficients[2])
result<-c(result,bostondis$coefficients[2])
result<-c(result,bostonrad$coefficients[2])
result<-c(result,bostontax$coefficients[2])
result<-c(result,bostonptratio$coefficients[2])
result<-c(result,bostonblack$coefficients[2])
result<-c(result,bostonlstat$coefficients[2])
result<-c(result,bostonmedv$coefficients[2])
display<-vector("numeric",0)
display<-c(display,bostonall$coefficients)
display<-display[-1]
plot(result,display,col='blue')

#the coefficient of the variables vary based on the modelling mode: alone or together

#Question 15D. Is there evidence of non-linear association between any of the predictors and the response?
bostonznnew<-lm(crim ~ poly(zn,3),data=Boston)
summary(bostonznnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonznnew)

bostonindusnew<-lm(crim ~ poly(indus,3),data=Boston)
summary(bostonindusnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonindusnew)

bostonnoxnew<-lm(crim ~ poly(nox,3),data=Boston)
summary(bostonnoxnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonnoxnew)

bostonrmnew<-lm(crim ~ poly(rm,3),data=Boston)
summary(bostonrmnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonrmnew)

bostonagenew<-lm(crim ~ poly(age,3),data=Boston)
summary(bostonagenew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonagenew)

bostondisnew<-lm(crim ~ poly(dis,3),data=Boston)
summary(bostondisnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostondisnew)

bostonradnew<-lm(crim ~ poly(rad,3),data=Boston)
summary(bostonradnew)


#Graph plot
plot(bostonradnew)


bostontaxnew<-lm(crim ~ poly(tax,3),data=Boston)
summary(bostontaxnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostontaxnew)

bostonptrationew<-lm(crim ~ poly(ptratio,3),data=Boston)
summary(bostonptrationew)

#Graph plot
plot(bostonptrationew)

bostonblacknew<-lm(crim ~ poly(black,3),data=Boston)
summary(bostonblacknew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonblacknew)

bostonlstatnew<-lm(crim ~ poly(lstat,3),data=Boston)
summary(bostonlstatnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonlstatnew)

bostonmedvnew<-lm(crim ~ poly(medv,3),data=Boston)
summary(bostonmedvnew)

#Graph plot
par(mfrow=c(2,2))
plot(bostonmedvnew)

#There is a nonlinear relationship between these variables: indus, nox, age, dis, ptratio, and medv. This relationship is cubic and for variable balck, the relationship is not cubic but non-linear.
```
